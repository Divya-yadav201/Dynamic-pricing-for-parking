```python
class ParkingSchema(pw.Schema):
  Timestamp: str
  Occupancy: int
  Capacity: int
  QueueLength: int
  TrafficConditionNearby: int
  IsSpecialDay: int
```

```python
data = pw.demo.replay_csv("parking_stream.csv", schema=ParkingSchema, input_rate=100)
```

```python
fmt = "%Y-%m-%d %H:%M:%S"

# Add new columns to the data stream:
# - 't' contains the parsed full datetime
# - 'day' extracts the date part and resets the time to midnight (useful for day-level aggregations)
data_with_time = data.with_columns(
    t = data.Timestamp.dt.strptime(fmt),
    day = data.Timestamp.dt.strptime(fmt).dt.strftime("%Y-%m-%dT00:00:00")
)
```

```python
import datetime

delta_window = (
  data_with_time.windowby(
      pw.this.t,
      instance = pw.this.day,
      window = pw.temporal.tumbling(datetime.timedelta(hours=3)),
      behavior = pw.temporal.exactly_once_behavior()
  )
  .reduce(
      t = pw.this._pw_window_end,
      occ_max = pw.reducers.max(pw.this.Occupancy),
      occ_min = pw.reducers.min(pw.this.Occupancy),
      cap = pw.reducers.max(pw.this.Capacity),
      qlen = pw.reducers.any(pw.this.QueueLength),
      traffic = pw.reducers.any(pw.this.TrafficConditionNearby),
      IsSpecialDay = pw.reducers.any(pw.this.IsSpecialDay)
  )
  .with_columns(
      price = 10 + (pw.this.occ_max - pw.this.occ_min) / pw.this.cap
              + 0.3 * pw.this.qlen
              + pw.this.traffic
              + 2 * pw.this.IsSpecialDay
  )
)
```

```python
import panel as pn
import bokeh
from bokeh.models import DatetimeTickFormatter

pn.extension()

# Step 1: Define your plotting function
def price_plotter(source):
    fig = bokeh.plotting.figure(
        height=400,
        width=800,
        title="Pathway: 3-Hourly Parking Price",
        x_axis_type="datetime",
    )
    fig.line("t", "price", source=source, line_width=2, color="navy")
    fig.scatter("t", "price", source=source, size=6, color="red", marker="circle")
    return fig

# Step 2: Create the dynamic plot
viz = delta_window.plot(price_plotter, sorting_col="t")

# Step 3: Helper to extract Bokeh figure safely
def find_bokeh_figure(layout):
    if hasattr(layout, "object") and hasattr(layout.object, "xaxis"):
        return layout.object
    elif hasattr(layout, "objects"):  # e.g., pn.Row or pn.Column
        for sub in layout.objects:
            fig = find_bokeh_figure(sub)
            if fig is not None:
                return fig
    elif isinstance(layout, (list, tuple)):
        for sub in layout:
            fig = find_bokeh_figure(sub)
            if fig is not None:
                return fig
    return None

# Step 4: Extract and customize
plot_figure = find_bokeh_figure(viz)

if plot_figure:
    plot_figure.xaxis.formatter = DatetimeTickFormatter(
        hours="%Y-%m-%d %H:%M",
        days="%Y-%m-%d",
        minutes="%H:%M"
    )
else:
    print(" Could not find Bokeh figure inside layout")

# Step 5: Display
pn.Column(viz).servable()
```
